```python
# importul bibliotecilor
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Importul datelor si vizualizarea lor

```python
# importul datelor
df=pd.read_csv('mouse_viral_study.csv')
```

```python
df.head()
```

```python
# vizuazlizarea gruprii claselor in funcie de caracteristici (scatterplot)
sns.scatterplot(data=df, x='Med_1_mL', y='Med_2_mL', hue='Virus Present')
```

```python
# crearea setului X si y
y=df['Virus Present']
X=df.drop('Virus Present', axis=1)
```

```python
#importul si crearea modelului SVC cu parametrii kernel='liniar' si C=1000
from sklearn.svm import SVC
model=SVC(kernel='linear', C=1000)
```

```python
#ajustarea modelului pe toate datele
model.fit(X, y)
```

```python
# importul functie de vizualizarea a hiperplanului si marginilor
from svm_margin_plot import plot_svm_boundary
```

```python
# vizualizarea datelor cu hiperplan si margini
plot_svm_boundary(model, X, y)
```

# 2. Analiza hiperparametrilor modelului SVC 

```python
# crearea mai multor modele cu diferite valori ale lui C si vizuazlizarea grafica a datelor
C=[0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10, 50, 100]
for i in C:
    model=SVC(kernel='linear', C=i)
    model.fit(X, y)
    plot_svm_boundary(model, X, y)
```

Hiper parametrul C determina numarul de puncte ce se vor afla intre marginile SVM.
Cu cresterea hipperparametrului C scade numarul punctelor ce se includ intre margini

```python
# crearea mai multor modele cu diferite valori posibile ale lui kernel
kernel=['linear', 'rbf', 'poly', 'sigmoid']
for i in kernel:
    model=SVC(kernel=i, C=1)
    model.fit(X, y)
    plot_svm_boundary(model, X, y)
```

Hiper parametrul kernel determina posibilitate de crearea a hiperplanului si a marginilor in functie de distinctia dintre clase.
Se observa ca atunci cand nu exista o distinctie buna intre clase cele mai bune rezultatele le asigura kernel='rbf' care este si valoarea implicita

```python
# crearea mai multor modele cu diferite valori posibile ale lui gamma
gamma=['scale', 'auto', 0.9, 2]
for i in gamma:
    model=SVC(kernel='rbf', C=1, gamma=i)
    model.fit(X, y)
    plot_svm_boundary(model, X, y)
```

Hiper parametrul gamma determina cum sunt influientate marginile de zgomot. Cel mai putin sunt influentate atunci cand gamma='scale' (valoarea implicita) si odata cu cresterea valorii creste influenta  

```python
# crearea mai multor modele cu diferite valori posibile ale lui degree pentru kernel='poly'
degree=[1, 2, 3, 4, 5]
for i in degree:
    model=SVC(kernel='poly', C=1, degree=i)
    model.fit(X, y)
    plot_svm_boundary(model, X, y)
```

Hiperparametrul degree determina neliniaritatea hiperplanului. Pentru degree=1 kernel poate fi considerat linear, iar cu cresterea lui degree creste neliniaritatea.

# 3. Elaborarea modelului SVC prin GridSearchCV

```python
# crearea dictionarului cu valori posibile pentru hiperparametri
param_grid={'C': [0.01, 0.1, 1, 10],
           'kernel':['linear', 'rbf', 'poly', 'sigmoid'],
           'gamma': ['scale', 'auto', 0.9, 2],
           "degree":[1, 2, 3, 4]}
```

```python
# crearea modelului SVC de baza cu valori implicite
model_SVC=SVC()
```

```python
# crearea modelului GridSearchCV
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(model_SVC, param_grid)
```

```python
# crearea seturilor de train si de test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=9)
```

```python
# Scalarea datelor
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scalat = scaler.fit_transform(X_train)
X_test_scalat = scaler.transform(X_test)
```

```python
# ajustarea modelului grid pe datele de training
grid.fit(X_train_scalat, y_train)
```

```python
# vizualizarea celor mai buni hiperparametri
grid.best_params_
```

```python
# realizarea predicitei pe datele de test
y_pred = grid.predict(X_test_scalat)
y_pred
```

```python
# determinarea acurateti modelului
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# determinarea confusion matrix a modelului
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

## 4.1 Modelul SVR de baza

```python
# importul datelor
df=pd.read_csv('cement_slump.csv')
```

```python
df.head()
```

```python
# vizualizarea corelatiei dintre coloanele df
plt.figure(figsize=(8,8), dpi=200)
sns.heatmap(df.corr(), annot=True);
```

```python
df.columns
```

```python
# crearea setului X si y
X = df.drop('Compressive Strength (28-day)(Mpa)', axis=1)
y = df['Compressive Strength (28-day)(Mpa)']
```

```python
# crearea seturilor de train si de test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=9)
```

```python
# Scalarea datelor
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scalat = scaler.fit_transform(X_train)
X_test_scalat = scaler.transform(X_test)
```

```python
# importul si crearea modelului CVR de baz
from sklearn.svm import SVR, LinearSVR
base_model=SVR()
```

```python
# ajustarea modelului de baza pe datele de training
base_model.fit(X_train_scalat, y_train)
```

```python
# realizarea predicitei pe datele de test
y_pred = base_model.predict(X_test_scalat)
y_pred
```

```python
# eroarea medie absoluta a predictiei
from sklearn.metrics import mean_absolute_error, mean_squared_error
mean_absolute_error(y_test, y_pred)
```

```python
# radicalul erorii medii patratice a predictiei
np.sqrt(mean_squared_error(y_test, y_pred))
```

```python
# valoarea medie a datelor de test
y_test.mean()
```
