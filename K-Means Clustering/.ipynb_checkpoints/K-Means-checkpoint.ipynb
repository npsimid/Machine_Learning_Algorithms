# 3. Crearea modelului K-Means

```python
# importul modulului KMeans from sklearn.cluster import KMeans
from sklearn.cluster import KMeans
```

```python
# crearea modelului pentru 2 clustere model = KMeans(n_clusters=2)
model = KMeans(n_clusters=2)
```

```python
# Ajustarea modelului și determinarea clusterului fiecarei date cluster_labels = model.fit_predict(X_scalat)
cluster_labels = model.fit_predict(X_scalat)
```

# 4. Valoarea optima a numarului de clustere

```python
#Crearea unei bucle for pentru determinarea valorilor sumei patratice a distantei pentru modele cu diferite valori 
# a numarului de clastere
suma = []

for k in range(2,10):
    model = KMeans(n_clusters=k)
    model.fit(X_scalat)
    suma.append(model.inertia_)
```

```python
# Vizualizarea valorilor sumelor patratice a distantei
suma
```

```python
# Vizualizarea grafica valorilor sumelor patratice a distantei plt.scatter(range(2,10),suma)
plt.scatter(range(2,10),suma)
plt.xlabel("Valoarea lui K Value")
plt.ylabel("Suma patratelor distantei");
```

# 5. Aplicarea K-Means la gruparea culorilor imagini

```python
# importul modulului matplotlib.image import matplotlib.image as mpimg
import matplotlib.image as mpimg
```

```python
# importul imaginii mpimg.imread('palm_trees.jpg')
imagine_array = mpimg.imread('palm_trees.jpg')
```

```python
# vizualizarea matricii imaginii
imagine_array
```

```python
# vizualizarea imaginii plt.imshow(imagine_array);
plt.figure(figsize=(6,6),dpi=200)
plt.imshow(imagine_array);
```

```python
# vizualizarea formei matricii imaginii imagine_array.shape
imagine_array.shape
```

```python
# citirea inaltimii, latimii si nuamrului de canale a imaginii h,w,c = imagine_array.shape
h,w,c = imagine_array.shape
```

```python
# transformarea matricii din 3d in 2d  imagine_array.reshape(h*w,c)
imagine_array2d=imagine_array.reshape(h*w,c)
```

```python
# crearea modelului K-Means pentru 6 clustere model = KMeans(n_clusters=6)
model = KMeans(n_clusters=6)
```

```python
# Ajustarea modelului și determinarea clusterului fiecarei date labels = model.fit_predict(imagine_array2d)
labels = model.fit_predict(imagine_array2d)
```

```python
# vizualizarea indicilor clusterelor
labels
```

```python
# determinarea coordonatelor centrelor clusterelor model.cluster_centers_
model.cluster_centers_
```

```python
# rotungirea si transformarea in valori intregi a valorilor coordonatelor centrelor clusterilor coduri_rgb = model.cluster_centers_.round(0).astype(int)
coduri_rgb = model.cluster_centers_.round(0).astype(int)
```

```python
# vizualizarea valorilor coordonatelor centrelor clusterilor
coduri_rgb
```

```python
# Crearea matricii imaginii cu culorile indentice pentru fiecare cluster imagine_cuatizata = np.reshape(coduri_rgb[labels], (h, w, c))
imagine_cuatizata = np.reshape(coduri_rgb[labels], (h, w, c))
```

```python
# vizualizarea matricii imaginii cuantizate
imagine_cuatizata
```

```python
# vizualizarea imaginii cuantizate plt.imshow(imagine_cuatizata)
plt.figure(figsize=(6,6),dpi=200)
plt.imshow(imagine_cuatizata)
```
