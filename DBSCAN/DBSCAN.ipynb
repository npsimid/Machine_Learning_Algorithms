# 2. Formarea clusterelor cu algoritmul K-Means

```python
# elaborarea unei functii de realizarea a predictiei si afisare a clusterelor
def afisare_clustere(model,data):
    labels = model.fit_predict(data)
    sns.scatterplot(data=data,x='X1',y='X2',hue=labels, palette='Set1')
```

```python
# importul modulului KMeans from sklearn.cluster import KMeans
from sklearn.cluster import KMeans
```

```python
# crearea modelului KMeans pentru 3 clustere model_K_3 = KMeans(n_clusters = 3)
model_K_3 = KMeans(n_clusters = 3)
```

```python
# apelarea functiei de realizarea predictie pe datele df1 afisare_clustere(model_K_3, df1)
afisare_clustere(model_K_3, df1)
```

```python
# crearea modelului KMeans pentru 2 clustere model_K_2 = KMeans(n_clusters = 2)
model_K_2 = KMeans(n_clusters = 2)
```

```python
# apelarea functiei de realizarea predictie pe datele df2 afisare_clustere(model_K_2, df2)
afisare_clustere(model_K_2, df2)
```

```python
# apelarea functiei de realizarea predictie pe datele df3
afisare_clustere(model_K_2, df3)
```

<a id='52'></a>
## 5.2. Valoarea optima in functie de numarul de outliers

```python
# crearea unei bucle de determinare a numarului de outliers pentru diferite valori a lui epsilon
numar_outliers= []
for i in np.linspace(0.001,10,100):
    model=DBSCAN(eps=i)
    model.fit(df5)
    n=np.sum(model.labels_ == -1)
    numar_outliers.append(n)
```

```python
# afisarea dependentei numarului de outliers de valoarea lui epsilon
sns.lineplot(x=np.linspace(0.001,10,100),y=numar_outliers)
plt.ylabel("Numarul de outliers")
plt.xlabel("Valoarea lui epsilon")
```

```python
# afisarea dependentei numarului de outliers de valoarea lui epsilon in gama de valori 0...1
sns.lineplot(x=np.linspace(0.001,10,100),y=numar_outliers)
plt.ylabel("Numarul de outliers")
plt.xlabel("Valoarea lui epsilon")
plt.xlim(0,1);
```

```python
# afisarea dependentei numarului de outliers in gama de valori 0...10 de valoarea lui epsilon 
sns.lineplot(x=np.linspace(0.001,10,100),y=numar_outliers)
plt.ylabel("Numarul de outliers")
plt.xlabel("Valoarea lui epsilon")
plt.ylim(0,10);
plt.hlines(y=3,xmin=0,xmax=10,colors='red',ls='--')
```

```python
# crearea modelului DBSCAN cu hiperparametrul epsilon de o valoare optima dbscan = DBSCAN(eps=0.7)
dbscan = DBSCAN(eps=0.7)
```

```python
# apelarea functiei de realizarea predictie pe datele df5
afisare_clustere(dbscan, df5)
```

<a id='53'></a>
## 5.3. Valoarea optima in functie de procentul de outliers

```python
# crearea unei bucle de determinare a procentajului de outliers pentru diferite valori a lui epsilon
procent_outliers= []
for i in np.linspace(0.001,10,100):
    model=DBSCAN(eps=i)
    model.fit(df5)
    p=100 * np.sum(model.labels_ == -1) / len(model.labels_)
    procent_outliers.append(p)
```

```python
# afisarea dependentei procentajului de outliers de valoarea lui epsilon
sns.lineplot(x=np.linspace(0.001,10,100),y=procent_outliers)
plt.ylabel("Procentajul de outliers")
plt.xlabel("Valoarea lui epsilon")
```

```python
# afisarea dependentei procentajului de outliers de valoarea lui epsilon in gama de valori 0...1
sns.lineplot(x=np.linspace(0.001,10,100),y=procent_outliers)
plt.ylabel("Procentajul de outliers")
plt.xlabel("Valoarea lui epsilon")
plt.xlim(0,1);
```

```python
# afisarea dependentei procentajului de outliers sub 1% de valoarea lui epsilon 
sns.lineplot(x=np.linspace(0.001,10,100),y=procent_outliers)
plt.ylabel("Procentajul de outliers")
plt.xlabel("Valoarea lui epsilon")
plt.ylim(0,1);
plt.hlines(y=0.5,xmin=0,xmax=10,colors='red',ls='--')
```

# 6. Valoarea optima a hiper-parametrului Minimum Samples

```python
# crearea modelului DBSCAN cu hiperparametrul min_samples de valoarea minima dbscan = DBSCAN(min_samples=1)
dbscan = DBSCAN(min_samples=1)
```

```python
# apelarea functiei de realizarea predictie pe datele df5
afisare_clustere(dbscan, df5)
```

```python
# crearea modelului DBSCAN cu setarea parametrilor eps si min_samples dbscan = DBSCAN(eps=0.7, min_samples=1)
dbscan = DBSCAN(eps=0.7, min_samples=1)
```

```python
# apelarea functiei de realizarea predictie pe datele df5
afisare_clustere(dbscan, df5)
```

```python
# crearea unei bucle de determinare a procentajului de outliers pentru diferite valori a numarului mimin
procent_outliers= []
for i in np.arange(1,100):
    model=DBSCAN(min_samples=i)
    model.fit(df5)
    p=100 * np.sum(model.labels_ == -1) / len(model.labels_)
    procent_outliers.append(p)
```

```python
# afisarea dependentei procentajului de outliers de valoarea lui min_samples
sns.lineplot(x=np.arange(1,100),y=procent_outliers)
plt.ylabel("Procentajul de outliers")
plt.xlabel("Numarul minim de puncte")
```

```python
# crearea modelului DBSCAN cu hiperparametrul min_samples de valoarea dubla numarului de caracteristici ale datelor
#dbscan = DBSCAN(min_samples=2*df5.shape[1])
dbscan = DBSCAN(min_samples=2*df5.shape[1])
```

```python
# apelarea functiei de realizarea predictie pe datele df5
afisare_clustere(dbscan, df5)
```
